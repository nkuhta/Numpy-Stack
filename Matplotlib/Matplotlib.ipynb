```python
import matplotlib.pyplot as plt
```

#  Line Plots

```python
import numpy as np
```

```python
x = np.linspace(0,10,10)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Sin Function")

plt.show()
```

```python
#  same plot with more x points 
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

#  Scatterplot

```python
import pandas as pd 
```

```python
A = pd.read_csv('data_1d.csv',header=None)
```

```python
A.head()
```

```python
#  turn df into a matrix 
A = A.as_matrix()
```

```python
x = A[:,0]
y = A[:,1]
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1
```

```python
plt.scatter(x,y)
plt.plot(x_line,y_line)
plt.show()
```
