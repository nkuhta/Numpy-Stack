#  Gaussian Probability Distribution Function (PDF)

$
\displaystyle
f\left(x\,\middle| \,\mu,\sigma \right)= \frac{1}{\sqrt{2 \sigma^2 \pi}} \exp\left(\frac{(x-\mu)^2}{2 \sigma^2}\right)
$

$
\mu = $ mean$
$

$
\sigma = $ standard deviation$
$

```python
from scipy.stats import norm
```

```python
#  standard normal pdf value at zero
norm.pdf(0)
```

```python
#  x = 0, mean = 5, sd = 10
norm.pdf(0,loc=5,scale=10)
```

```python
import numpy as np
```

```python
r = np.random.randn(10)
```

```python
#  calculate the pdf of all r values at once 
norm.pdf(r)
```

#  Log PDF

```python
norm.logpdf(r)
```

#  Cummulative Distribution Function (CDF)

$
\displaystyle
CDF = \int\limits_{-\infty}^{X}  PDF
$

*  Note this must be done numerically 

```python
norm.cdf(r)
```

```python
norm.logcdf(r)
```

#  General Multivariate Normal Sampling

```python
cov = np.array([[1,0.8],[0.8,3]])
```

```python
from scipy.stats import multivariate_normal as mvn
```

```python
mu = np.array([0,2])
```

```python
r = mvn.rvs(mean=mu,cov=cov,size=1000)
```

```python
plt.scatter(r[:,0],r[:,1],edgecolor='black')
plt.axis('equal')
plt.show()
```

#  Fourier Transform (FFT)

```python
x = np.linspace(0,100,10000)
```

```python
y = np.sin(x) + np.sin(3*x) + np.sin(5*x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
yfft = np.fft.fft(y)
```

```python
plt.plot(np.abs(yfft))
plt.xlim(0,100)
plt.show()
```
